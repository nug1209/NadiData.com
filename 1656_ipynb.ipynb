```python
import pandas as pd
```

```python
df = pd.read_table('data.txt')
df.head()
```

```python
df = df.drop(['Notes', 'Year Code'], axis=1)
df.head()
```

```python
df = df.head(2701)
df.tail()
```

```python
df["Mother's Bridged Race"].unique()
```

```python
df = df.astype({'State Code': int,
               'Year': int,
               'Births': int})

df.head()
```

```python
# df['weight'] / df.groupby('word')['weight'].transform('sum')

df["Births Normalized"] = (df['Births'] - df.groupby(by=["State", "Mother's Bridged Race"])["Births"].transform('min')) / (df.groupby(by=["State", "Mother's Bridged Race"])["Births"].transform('max') - df.groupby(by=["State", "Mother's Bridged Race"])["Births"].transform('min'))
df.head()
```
